```python
import trimesh
import mesh2sdf
import numpy as np
import time
import generationtools as gnt

size = 64

mesh = trimesh.load(meshPath)
t0 = time.time()
sdf,mesh = gnt.mesh_to_sdf_tensor(mesh, size, recenter=False)
t1 = time.time()
fixedMeshPath =  meshPath[:-4] + "_fixed.obj"
sdfPath = meshPath[:-4] + ".npy"
tfPath = meshPath[:-4] + ".pt"
# output
mesh.export(fixedMeshPath)
np.save(sdfPath, sdf)
print('It takes %.4f seconds to process' % (t1-t0))
mesh.show()
```

```python
mesh.export("data/Paper_examples/Sofa_sdf.obj")
```

```python
import torch
shapedData = np.reshape(sdf, [1, 1, 64, 64, 64])
torch.save(torch.from_numpy(shapedData[:,:,:64,:64,:64]), tfPath)
```
